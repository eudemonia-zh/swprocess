```python
import math

import matplotlib.pyplot as plt
import numpy as np

import swprocess
```

## Input
---

### Files

```python
# File(s) to import in the .max format produced using Geopsy and spac-style processing.
fnames = ["data/spac/nz_wghs_c50_spac.max",
          "data/spac/nz_wghs_bigx_spac.max"]
```

### Component

```python
# Component {0 - vertical, 1 - radial, 2 - transverse}
component = 0
```

### Plot Settings

```python
# Velocity search limits for fitting experimental spac curves.
vrange = (150, 1000)

# Domains in which to plot the experimental dispersion data.
domains = [
    ["frequency", "velocity"],
    ["wavelength", "velocity"],
]

# Colors one per `fnames_set` entry. Examples include: "limegreen", "dodgerblue", "tomatoe", "darkorange".
# A full listing is provided here: https://matplotlib.org/3.1.0/_images/sphx_glr_named_colors_003.png
# Colors may also be listed in hexidecimal.
colors = ["dodgerblue", "tomato"]

# Custom labels one per `fanems_set` entry. Note these will also be used for the output file suffixs.
labels = ["c50", "bigx"]
```

## Plot Experimental SPAC Ratios
---

```python
spaccurvesuites = []
for fname, color in zip(fnames, colors):
    print(f"Loading: {fname}")
    spaccurvesuite = swprocess.SpacCurveSuite.from_max(fname=fname, component=f"({component})")
    spaccurvesuites.append(spaccurvesuite)
    
    rings = spaccurvesuite.rings
    
    ncols = min(len(rings), 3)
    nrows = max(math.ceil(len(rings)/ncols), 1)
    
    fig, axs = plt.subplots(ncols=ncols, nrows=nrows, sharey=True, sharex=True, figsize=(9,3*nrows), dpi=150, gridspec_kw=dict(wspace=0.1, hspace=0.07))
    
    for ax, ring in zip(axs.flatten(), rings):
        for spaccurve in spaccurvesuite.spaccurves:
            if spaccurve.ring == ring:
                ax.plot(spaccurve.frequencies, spaccurve.ratios, marker=".", markersize=1, linestyle="")
    plt.show()

#         ax.text(0.95, 0.95, f"ring{ring}: {spaccurve.dmin:.1f}-{spaccurve.dmax:.1f} m", ha="right", va="top", transform=ax.transAxes)
```

## Plot Dispersion Data
---

```python
xtype = [x for x, _ in domains]
ytype = [y for _, y in domains]
fig, axs = plt.subplots(ncols=len(xtype), figsize=(9,3), dpi=150, gridspec_kw=dict(wspace=0.4))

if len(fnames) != len(colors) or len(fnames) != len(labels):
    raise IndexError(f"fnames_set, colors, and labels must be the same length.")

# suites = []
for spaccurvesuite, color, label in zip(spaccurvesuites, colors, labels):
    peaksuite_byring = spaccurvesuite.to_peaksuite(vrange=vrange)

    for ring, peaksuite in zip(spaccurvesuite.rings, peaksuite_byring):
        peaksuite.plot(xtype=xtype, ax=axs, ytype=ytype, plot_kwargs=dict(color=color, label=label))
#         suites.append(peaksuite)

axs[-1].legend(bbox_to_anchor = (1.1, 0.5), loc="center left")

plt.show()
```
